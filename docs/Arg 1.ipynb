# Industry, infrastructure, and nature.

With industry it is understandable that plenty of advanced infrastructure must follow. Things such as airports, regular ports, train stations and railroads, motorways, and roads are all natural results of an advancing society. These improvements in local infrastructure are a direct sign of an increase in wealth since you would rarely see dirt and gravel roads in an advanced first world country. As more and more infrastructure is installed within a country said country often finds itself in the position of being able to shape the rest of its land in whatever way it pleases bringing plenty of other features to its nation.

The opposite is then also true, with more roads less money and land is being spent on the preservation of nature.
